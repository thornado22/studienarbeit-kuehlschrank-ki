```python
import matplotlib.pyplot as plt
import seaborn as sns

def plot_f1(x, y, title, path, figsize=(8, 6), palette="viridis", fmt="{:.2f}"):
    # load json

    # create bar plot
    plt.figure(figsize=figsize)
    ax = sns.barplot(x=x, y=y, palette=palette)

    ax.set_xlabel("Klasse")
    ax.set_ylabel("F1 Score")
    ax.set_title(f"Validierungs-Klassen-F1-Score aus {title}")

    plt.xticks(rotation=45, ha="right")

    # annotate bars with values
    for i, v in enumerate(y):
        ax.text(i, v, fmt.format(v), ha="center", va="bottom", fontsize=9)

    plt.tight_layout()
    plt.show()
    plt.savefig(path)
```

```python
from stats import plot_f1
import json

file = "../models/test_3.json"

with open(file, "r") as f:
    src = json.load(f)

val_class_f1 = src["results"]["val_class_f1"]
classes = src["meta"]["classes"]

plot_f1( classes, val_class_f1, "1", "f1_test3.png")
```
