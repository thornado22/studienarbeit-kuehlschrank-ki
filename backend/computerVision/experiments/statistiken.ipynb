```python
from loader import LabelLoader, MetaLoader
label = LabelLoader("./images/data/labels.json").load(0)
meta = MetaLoader("./images/umfrage_antworten_11_05.csv", "./images/data/labels.json").add_food_type("./stats/mapping_lebensmittelarten.json", False)
meta = meta.drop(columns=["Zeitstempel"])
meta = meta.rename(columns={'pescetarisch (vegetarisch + Fisch)': 'pescetarisch'})
```

```python
from stats.stats import create_barchart


anzahl_personen = ['1', '2', '3', '4', '5', '6 +']
geschlecht = ['divers', 'männlich', 'weiblich']
alter =  ['18 - 25', '25 - 35', '35 - 60', 'über 60']
ernaerung = ['glutenfrei', 'halal', 'keine Einschränkungen', 'laktosefrei', 'pescetarisch', 'vegan', 'vegetarisch']
inhalt = ['obst', 'gemuese', 'milch_produkte', 'fleisch_und_alternativen', 'fisch', 'fertig_produkte', 'sonstiges']

create_barchart(meta, "Angegebenes Alter der einkaufenden Person", "Alter", "Anzahl", cols=alter)
create_barchart(meta, "Angegebene Anzahl der Kühlschranknutzer", "Nutzer", "Anzahl", cols=anzahl_personen)
create_barchart(meta, "Angegebenes Geschlecht der einkaufenden Person", "Geschlecht", "Anzahl", cols=geschlecht)
create_barchart(meta, "Angegebene Ernährungsweise", "Ernährungsweise", "Anzahl", cols=ernaerung)
create_barchart(meta, "Inhalt des Kühlschranks", "Lebensmittelgruppe", "Anzahl", cols=inhalt)
```
