```python
from computerVision.loader import LabelLoader, MetaLoader
from computerVision.stats.label_relation_toolbox import cluster_labels

from pathlib import Path

ROOT = Path.cwd()
while not (ROOT / "computerVision").exists():
    ROOT = ROOT.parent

label_json = str(ROOT / "computerVision/images/data/labels.json")
mapping = str(ROOT / "computerVision/stats/mapping_lebensmittelarten.json")
survey_results = str(ROOT / "computerVision/images/umfrage_antworten_11_05.csv")

labelLoader = LabelLoader(label_json)
```

## Clustering der Labels

- Labels müssen min. 15 mal vorkommen
- keine Metadaten

```python

label_df = labelLoader.load(15, False)


results = cluster_labels(
    label_df,
    n_components_umap=10,
    min_cluster_size=15,
    min_samples=5,
    plot=True
)
```

- Labels müssen min 15 mal vorkommen
- mit Metadaten

```python
label_df = labelLoader.load(15, True)
    

results = cluster_labels(
    label_df,
    n_components_umap=10,
    min_cluster_size=15,
    min_samples=5,
    plot=True
)
```

## Clustering der Metadaten

```python
meta_loader = MetaLoader(survey_results, label_json)
meta = meta_loader.load()
meta = meta.drop("Zeitstempel", axis=1)

results = cluster_labels(
    meta,
    n_components_umap=10,
    min_cluster_size=15,
    min_samples=5,
    plot=True
)
```

- mit den Lebensmittelarten

```python
food_type = meta_loader.add_food_type(mapping, False)
food_type = food_type.drop("Zeitstempel", axis=1)

results = cluster_labels(
    food_type,
    n_components_umap=10,
    min_cluster_size=15,
    min_samples=5,
    plot=True
)
```
